```python
import os
import cv2
import numpy as np

# Define paths
image_dir = "/home/marci/Documents/gits/persuasion-techniques-in-memes/team_2/data/processed/data_kaist"
txt_dir = "/home/marci/Documents/gits/persuasion-techniques-in-memes/team_2/data/processed/data_kaist_yolo"
output_dir = "/home/marci/Documents/gits/persuasion-techniques-in-memes/team_2/data/processed/data_2"

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Iterate through all .jpg images in the image directory
for filename in os.listdir(image_dir):
    if filename.endswith('.jpg'):
        # Construct the full path to the image and the corresponding .txt file
        image_path = os.path.join(image_dir, filename)
        txt_path = os.path.join(txt_dir, filename.replace('.jpg', '.txt'))

        # Load the image
        image = cv2.imread(image_path)
        if image is None:
            print(f"Warning: could not read {image_path}")
            continue
        
        height, width = image.shape[:2]

        # Check if the .txt file exists
        if not os.path.isfile(txt_path):
            print(f"Warning: no corresponding .txt file found for {filename}")
            continue

        # Open and read bounding box data from the .txt file
        with open(txt_path, 'r') as file:
            for line in file:
                parts = line.strip().split()
                if len(parts) == 5:
                    # Parse bounding box coordinates from the .txt file
                    x_center = float(parts[1]) * width
                    y_center = float(parts[2]) * height
                    box_width = float(parts[3]) * width
                    box_height = float(parts[4]) * height

                    # Convert YOLO format to rectangle coordinates
                    x1 = int(x_center - box_width / 2)
                    y1 = int(y_center - box_height / 2)
                    x2 = int(x_center + box_width / 2)
                    y2 = int(y_center + box_height / 2)

                    # Draw the rectangle on the image (red color, thickness 2)
                    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 0, 255), 2)
                else:
                    print(f"Warning: line format incorrect in {txt_path}")
        
        # Save the modified image to the output directory
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, image)
        print(f"Processed and saved: {output_path}")
```
